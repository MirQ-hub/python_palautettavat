```python
import pandas as pd

df=pd.read_csv(r"c:\2022-2023\PYTHON, ville majava\titanic_for_exercise.csv")
print("Initial dataframe:")
df.head()
```

```python
#Check missing values
print("Missing values")
print(df.isnull)
```

```python
#find index of rows with errornous values
index_names = df[(df["Survived"] !=0)&(df["Survived"] !=1)].index

df.drop(index_names, inplace=True)
print(df["Survived"].unique())
print(df)
```

```python
#Check missing values
index_names=df[(df["Sex"] !="male") & (df["Sex"] !="female")].index
print(index_names)
df.drop(index_names, inplace=True)
print(df["Sex"].unique())
print(df)
```

```python
#find index of rows with errornous values
index_names = df[(df["Age"] ==0)].index

df.drop(index_names, inplace=True)

print(df)
```
