```python
from dash import Dash, html, dcc, Input, Output
from dash.dependencies import Input, Output
import pandas as pd
import plotly.express as px

results = {'Hypothesis 1': {'t-test': [male, female], 'samples data': [male, female]}}


app = Dash(__name__)

app.layout = html.Div([
    html.H1("Powerlifting data analysis", style={'textAlign': 'center'}),
    dcc.Dropdown(options=[{'label': key, 'value': key} for key in list(results.keys())], value='Hypothesis 1', id='dropdown1'),
    dcc.Graph(id='graph1')
])

@app.callback(Output('graph1', 'figure'), [Input('dropdown1', 'value')])
def update_graph(selected_dropdown_value):
    dict_res = results[selected_dropdown_value]
    df_data = pd.DataFrame(dict_res['samples data']).transpose() 
    
    # Add t-test results to the figure title

    df_t_test = pd.DataFrame(dict_res['t-test']).transpose() 
    g = px.histogram(df_data, x=df_data.columns, barmode='overlay', color_discrete_sequence=['blue', 'red'])
    g.update_layout(title_text="t-statistics = " + str(df_t_test[0][0]) + ", p-value = " + str(df_t_test[0][1]))
    return g

if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
#Hypothesis 4:"Men lift more than women when bodyweight is 100-105kg"


import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

male = df["Best3BenchKg"].loc[(df["Sex"] == "M") & (df["Event"] == "SBD") & 
                              (df["Equipment"] == "Raw") & (df["Federation"] == "IPF") & 
                              (df["BodyweightKg"].between(100, 105))].dropna().to_list()

female = df["Best3BenchKg"].loc[(df["Sex"] == "F") & (df["Event"] == "SBD") & 
                                (df["Equipment"] == "Raw") & (df["Federation"] == "IPF") & 
                                (df["BodyweightKg"].between(100, 105))].dropna().to_list()

# Plotting data
plt.figure(figsize=(10, 6))
sns.histplot(male, color="blue", label="Male")
sns.histplot(female, color="red", label="Female")
plt.xlabel('Total Kilograms')
plt.legend()

# Performing t-test
t, p = stats.ttest_ind(male, female, equal_var=False)

print("T-statistics = " + str(t))
print("P-value = " + str(p))
print("Difference of mean values = " + str(np.mean(male) - np.mean(female)) + ' Kg')
print("Number of samples: " + str(len(male)) + " and " + str(len(female)))

# Conclusion of t-test
if p < 0.05:
    print("We reject the null hypothesis. There is a significant difference between the means of the two groups.")
else:
    print("We accept the null hypothesis. There is no significant difference between the means of the two groups.")

# Update results dictionary with Hypothesis 4
results.update({'Hypothesis 4': {'t-test': [t, p], 'samples data': [male, female]}})

```

```python
# Hypothesis 5: "Men lift more in IPF than in WPC" compare "TotalKg" lifts

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

IPF = df["TotalKg"].loc[(df["Sex"] == "M") & (df["Event"] == "SBD") & 
                        (df["Equipment"] == "Raw") & (df["Federation"] == "IPF")].dropna().to_list()

WPC = df["TotalKg"].loc[(df["Sex"] == "M") & (df["Event"] == "SBD") & 
                        (df["Equipment"] == "Raw") & (df["Federation"] == "WPC")].dropna().to_list()

# Plotting data
plt.figure(figsize=(10, 6))
sns.histplot(IPF, color="blue", label="IPF")
sns.histplot(WPC, color="red", label="WPC")
plt.xlabel('Total Kilograms')
plt.legend()

# Perform independent two-sample t-test
t, p = stats.ttest_ind(IPF, WPC, equal_var=False)

print("T-statistics = " + str(t))
print("P-value = " + str(p))
print("Difference of mean values = " + str(np.mean(IPF) - np.mean(WPC)) + ' Kg')
print("Number of samples: " + str(len(IPF)) + " and " + str(len(WPC)))

# Conclusion of t-test
if p < 0.05:
    print("We reject the null hypothesis. There is a significant difference between the means of the two groups.")
else:
    print("We accept the null hypothesis. There is no significant difference between the means of the two groups.")

# Update results dictionary with Hypothesis 5
results.update({'Hypothesis 5': {'t-test': [t, p], 'samples data': [IPF, WPC]}})
```

```python
#Hypothesis 6:"Lifter country affect to TotalKg"

import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import f_oneway

# Extract relevant columns from the DataFrame
country_total_kg = df[['Country', 'TotalKg']].dropna()

# Visualization - Box plot
plt.figure(figsize=(12, 6))
sns.boxplot(data=country_total_kg, x='Country', y='TotalKg')
plt.xlabel('Country')
plt.ylabel('Total Kilograms')
plt.title('Total Kilograms Lifted by Country')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Statistical Testing - ANOVA
countries = country_total_kg['Country'].unique()
country_groups = [country_total_kg[country_total_kg['Country'] == country]['TotalKg'] for country in countries]
f_statistic, p_value = f_oneway(*country_groups)

print("ANOVA Results:")
print("F-statistic:", f_statistic)
print("p-value:", p_value)

# Conclusion
if p_value < 0.05:
    print("We reject the null hypothesis. There is a significant difference in total kilograms lifted among different countries.")
else:
    print("We accept the null hypothesis. There is no significant difference in total kilograms lifted among different countries.")
```
